#### 1. Input file: Mass ratio

```python
InputFile_q = "Mass_ratio.dat"
Data_q   = np.genfromtxt(InputFile_q)


DecYear = Data_q[:,0]
q = Data_q[:,1]
dq = Data_q[:,2]
```

```python
fig, (ax0) = plt.subplots(1, 1, sharex=True, sharey=False, figsize=(10, 5), tight_layout=True)
#plt.xlim(BJD_time[0], BJD_time[-1])
plt.xlabel("Year")

ax0.tick_params(direction='in', which='both', bottom='on',top='on', right = 'on')
ax0.grid(linestyle='dotted')

ax0.errorbar(
DecYear[0], q[0], yerr=dq[0], fmt='o', color='black',alpha = 1.0,
                    ecolor='lightgray', elinewidth=2, capsize=0, markersize='6.00', label = 'Bailey et al. (1993)')

ax0.errorbar(
DecYear[1], q[1], yerr=dq[1], fmt='s', color='black',alpha = 1.0,
                    ecolor='lightgray', elinewidth=2, capsize=0, markersize='6.00', label = 'Pandel et al. (2002)')

ax0.errorbar(
DecYear[2:26], q[2:26], yerr=dq[2:26], fmt='D', color='black',alpha = 1.0,
                    ecolor='lightgray', elinewidth=2, capsize=0, markersize='5.00', label = 'This work')
ax0.legend(loc="best")
ax0.set_ylabel("Mass ratio")

#plt.text(No_q[0], 0.165, '1993')
#plt.text(No_q[1], 0.165, '2002')
#plt.text(No_q[2], 0.165, '2014')
#plt.text(No_q[3], 0.165, '2015')
#plt.text(No_q[4], 0.165, '2020')

fig.align_ylabels()
#output_filename = os.path.splitext(__file__)[0] + '.png'
#plt.savefig(output_filename, dpi=1000)
plt.savefig("2022_Feb_week1_lcurve_plot_Mass_ratio.jpg", format='jpg', dpi=1200)
plt.show()
```
